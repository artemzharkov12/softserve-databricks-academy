Creating a data with iteresting columns for us

id (GlobalEventID): A globally unique identifier assigned to the event by the GDELT system.

EventDate (SqlDate): The date the event took place, parsed into a standard YYYY-MM-DD Date format for time-series analysis.

Actor1 (Actor1Name): The initiator or active subject of the news event (e.g., "SEC", "ELON MUSK", "CHINA").

Actor1Country (Actor1CountryCode): The 3-letter CAMEO country code associated with Actor 1 (e.g., "USA", "CHN").

Actor2 (Actor2Name): The receiver or target of the action described in the news event.

Actor2Country (Actor2CountryCode): The 3-letter CAMEO country code associated with Actor 2.

**EventCode: **The standardized CAMEO classification code that describes the specific type of action (e.g., 010 for "Make statement", 190 for "Use conventional force").

GoldsteinScale: A numeric score ranging from -10.0 to +10.0 that measures the theoretical impact of the event on global stability (negative for conflict, positive for cooperation).

AvgTone: The average sentiment score of all news coverage regarding this event. Usually ranges from -10 (very negative) to +10 (very positive), helping to identify market fear or greed.

Mentions (NumMentions): The total number of times this specific event was mentioned across the global GDELT monitoring network.

NumArticles: The total number of unique news articles that covered this event. Great for filtering out local noise and keeping only globally impactful news.

SourceUrl: A direct web link to the original news article that reported the event.

```python
df_selected_column = df_bronze.select(
    col("GlobalEventID").alias("id"),
    to_date(col("SqlDate"), "yyyyMMdd").alias("EventDate"),

    col("Actor1Name").alias("Actor1"),
    col("Actor1CountryCode").alias("Actor1Country"),
    col("Actor2Name").alias("Actor2"),
    col("Actor2CountryCode").alias("Actor2Country"),
    col("EventCode"),        
    col("GoldsteinScale"),   
    col("AvgTone"),          
    col("NumMentions").alias("Mentions"),     
    col("NumArticles"),
    col("SourceUrl")
)
```
