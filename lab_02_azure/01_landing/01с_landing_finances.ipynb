```python
import yfinance as yf
import pandas as pd
import os
from datetime import datetime
```

```python
VOLUME_PATH = "/Volumes/dbr_dev/artemzharkov10_bronze/raw_data/finance_history/"
try:  
    os.makedirs(VOLUME_PATH, exist_ok=True)
except Exception as e:
    print(f"Could not create directory {VOLUME_PATH}: {e}")
```

```python

TARGETS = {
    "BITCOIN": "BTC-USD",
    "GOLD": "GC=F",
    "OIL": "CL=F"
}
START_DATE = "2015-01-01"
END_DATE = datetime.today().strftime("%Y-%m-%d")

for fin,fin_code in TARGETS.items():
    try:
        print(f"Downloading {fin}")
        file_path = os.path.join(VOLUME_PATH, f"{fin}.csv")
        if os.path.exists(file_path):
            print(f"File {file_path} already exists. Skip -> ")
            continue
        # download data via yfinance
        df = yf.download(fin_code, start=START_DATE, end=END_DATE)
        if df.empty:
            print(f"Empty dataframe for {fin}.")
        else:
            #clearing of columns
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = df.columns.droplevel(1)
            df = df.reset_index()
            df.rename(columns={'Price': 'Date'}, inplace=True)

            df.to_csv(file_path, index=False) #index = false ? 
    except Exception as e:
        print(f"Could not download {fin}: {e}")



```
