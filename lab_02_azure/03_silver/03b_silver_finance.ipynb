```python
from pyspark.sql.functions import col, to_date, to_timestamp,current_timestamp,round,current_date
```

```python
dbutils.widgets.text(
    "bronze_catalog",
     "dbr_dev")
dbutils.widgets.text(
    "bronze_schema",
    "artemzharkov10_bronze"
)

dbutils.widgets.text(
    "silver_catalog",
     "dbr_dev")
dbutils.widgets.text(
    "silver_schema",
    "artemzharkov10_silver"
)

BRONZE_CATALOG = dbutils.widgets.get("bronze_catalog")
BRONZE_SCHEMA = dbutils.widgets.get("bronze_schema")

SILVER_CATALOG = dbutils.widgets.get("silver_catalog")
SILVER_SCHEMA = dbutils.widgets.get("silver_schema")
    
```

```python
# Bitcoin

df_bronze = spark.read.table(f"{BRONZE_CATALOG}.{BRONZE_SCHEMA}.finance_bitcoin_bronze")
df_clean = df_bronze.dropna(subset = ["Date","Close","Open","High","Low","Volume"])

df_transform = (
    df_clean
    .withColumnRenamed("Date", "TradeDate")
    .withColumn("TradeDate", to_timestamp(col("TradeDate"), "yyyy-MM-dd"))
    .withColumn("Close", round(col("Close").cast("double"), 2))
    .withColumn("Open", round(col("Open").cast("double"), 2))
    .withColumn("High", round(col("High").cast("double"), 2))
    .withColumn("Low", round(col("Low").cast("double"), 2))
    .withColumn("Volume", col("Volume").cast("double"))
    # meta
    .withColumn("source_filename", col("_metadata.file_path"))
    .withColumn("ingest_timestamp", current_timestamp())
    .withColumn("load_date", current_date())
)

df_silver = df_transform.dropDuplicates(["TradeDate"])

df_silver.write.mode("overwrite").saveAsTable(f"{SILVER_CATALOG}.{SILVER_SCHEMA}.finance_bitcoin_silver")
# print(df_silver.count())
# display(df_silver.limit(20))
```
