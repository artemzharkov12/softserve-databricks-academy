```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, DateType
```

```python
SOURCE_VOLUME_PATH = "/Volumes/dbr_dev/artemzharkov10_bronze/raw_data/finance_history/BITCOIN.csv"

market_schema = StructType([
    StructField("Date", DateType(), True), 
    StructField("Close", DoubleType(), True),
    StructField("High", DoubleType(), True),
    StructField("Low", DoubleType(), True),
    StructField("Open", DoubleType(), True),
    StructField("Volume", DoubleType(), True)
])
```

```python

df_bronze = (spark.read.format("csv")
  .option("header", "true")
  .schema(market_schema)
  .load(SOURCE_VOLUME_PATH)) 

(df_bronze.write
  .format("delta")
  .mode("overwrite")
  .option("overwriteSchema", "true")
  .saveAsTable("dbr_dev.artemzharkov10_bronze.finance_bitcoin_bronze"))

display(df_bronze.limit(30))
```
